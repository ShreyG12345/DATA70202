```python
# -*- coding: utf-8 -*-
"""
Spyder Editor

This is a temporary script file.
"""

#PLEASE NOTE: This code is SENSITIVE and must not be shared publicly, it is only for viewing by module convenors

# Load Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Making a Colour Pallette
colours = ['#377eb8', '#ff7f00', '#4daf4a',
                  '#f781bf', '#a65628', '#984ea3',
                  '#999999', '#e41a1c', '#dede00']

# Load plots in new window
# %matplotlib qt

# Load Data
#NOTE - the file loaded below is a proxy name to protect sensitive data as per GAD's request
df = pd.read_csv(r'D:\UoM 2025 - Gender Pay and Pension Analysis\0. Raw Data - DO NOT COPY\processed_data.csv')

# Dataset Info
df.columns
df.dtypes

print('Basic Info: ')
print(df.info())
print("\nSummary Statistics (Numerical):")
print(df.describe().T)
print("\nSummary Statistics (Categorical):")
print(df.describe(include=['object']).T)

# Missing Values
print(df.isnull().sum().sort_values(ascending=False))
print(df.isnull().mean() * 100)


#%%
# Categorical Columns
# Gender
# Make everything is lowercase and get rid of spaces
df['gender'] = df['gender'].str.strip().str.lower()

# Cleaning Variables
df['gender'] = df['gender'].replace({
    'm': 'Male',
    'male': 'Male',
    'f': 'Female',
    'female': 'Female'
})

# Summary Statistics
df['gender'].value_counts()
df['gender'].value_counts(normalize=True)


# Visualisation
gender_counts = df['gender'].value_counts()

plt.figure(figsize=(8, 5))

bars = plt.bar(gender_counts.index, gender_counts.values, color=colours[:len(gender_counts)])

plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')

for i, v in enumerate(gender_counts.values):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom', fontsize=10)

plt.show()

#%%
# Marital Status
# Cleaning Variables
df['marital_status'] = df['marital_status'].str.upper()

# Summary Statistics
df['marital_status'].value_counts()
df['marital_status'].value_counts(normalize=True)

marital_labels = {
    'S': 'Single',
    'M': 'Married',
    'U': 'Unknown',
    'D': 'Divorced / Dissolved Partnership',
    'P': 'Declared Partnership',
    'C': 'Civil Partnership',
    'W': 'Widow/Widower',
    'B': 'Unspecified'
    }

marital_counts = df['marital_status'].value_counts()
marital_counts.index = marital_counts.index.map(marital_labels)

# Visualisation
plt.figure(figsize=(10, 6))

bars = plt.bar(marital_counts.index, marital_counts.values, color=colours[:len(marital_counts)])

plt.title('Marital Status Distribution')
plt.xlabel('Marital Status')
plt.ylabel('Count')

plt.xticks(rotation=30, ha='right')

for i, v in enumerate(marital_counts.values):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

#%%
# Rank
df['rank'].value_counts()
df['rank'].value_counts(normalize=True)

print("Unique Values:", df['rank'].unique())
print("Missing values:", df['rank'].isna().sum())

#dropping the NA's
df_rank_clean = df['rank'].dropna().astype(str)
rank_counts = df_rank_clean.value_counts().sort_index()

# Keeping the NA's
rank_counts = df['rank'].fillna('Missing').astype(str).value_counts().sort_index()

#Visualisation
plt.figure(figsize=(10, 6))
bars = plt.bar(rank_counts.index, rank_counts.values, color=colours[:len(rank_counts)])

plt.title('Rank Distribution')
plt.xlabel('Rank')
plt.ylabel('Count')
plt.xticks(rotation=30, ha='right')

for i, v in enumerate(rank_counts.values):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

df['rank'] = df['rank'].fillna('Missing').astype(str)
df['rank'].value_counts()
df['rank'].value_counts(normalize=True)

#%%
#current scheme
print("Unique Values:", df['current_scheme'].unique())
print("Missing values:", df['current_scheme'].isna().sum())

df['current_scheme'].value_counts()
df['current_scheme'].value_counts(normalize=True)

current_scheme_counts = df['current_scheme'].value_counts()
plt.figure(figsize=(8, 5))
bars = plt.bar(current_scheme_counts.index, current_scheme_counts.values, color=colours[:len(current_scheme_counts)])
plt.title('Current Scheme Distribution')
plt.xlabel('Scheme')
plt.ylabel('Count')
for i, v in enumerate(current_scheme_counts.values):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom', fontsize=10)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
#plt.show()



#%%
# Part Time
df['part_time'].value_counts()
df['part_time'].value_counts(normalize=True)

# Missing Values
print("Missing values:", df['part_time'].isna().sum())

df['part_time'] = df['part_time'].fillna('N')
df['part_time'].value_counts()
df['part_time'].value_counts(normalize=True)

part_time_counts = df['part_time'].value_counts()
plt.figure(figsize=(6, 4))
bars = plt.bar(part_time_counts.index, part_time_counts.values, color=colours[:len(part_time_counts)])
plt.title('Part-Time Status Distribution')
plt.xlabel('Part-Time')
plt.ylabel('Count')
for i, v in enumerate(part_time_counts.values):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom', fontsize=10)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()


#%%
#career break
df['career_break'] = df['career_break'].replace({
    'y': 'Y'
})
df['career_break'] = df['career_break'].fillna('N')

df['career_break'].value_counts()
df['career_break'].value_counts(normalize=True)

career_break_counts = df['career_break'].value_counts()
plt.figure(figsize=(6, 4))
bars = plt.bar(career_break_counts.index, career_break_counts.values, color=colours[:len(career_break_counts)])
plt.title('Career Break Status')
plt.xlabel('Career Break')
plt.ylabel('Count')
for i, v in enumerate(career_break_counts.values):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom', fontsize=10)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()


#%%
#protection status
df['protection_status'].value_counts()
df['protection_status'].value_counts(normalize=True)

protection_status_counts = df['protection_status'].value_counts()
plt.figure(figsize=(8, 5))
bars = plt.bar(protection_status_counts.index, protection_status_counts.values, color=colours[:len(protection_status_counts)])
plt.title('Protection Status Distribution')
plt.xlabel('Protection Status')
plt.ylabel('Count')
for i, v in enumerate(protection_status_counts.values):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom', fontsize=10)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()


#%%
#Previous final salary scheme orig
df['previous_final_salary_scheme_orig'].value_counts()
df['previous_final_salary_scheme_orig'].value_counts(normalize=True)

df['previous_final_salary_scheme_orig'] = df['previous_final_salary_scheme_orig'].fillna('N')
df['previous_final_salary_scheme_orig'].value_counts()
df['previous_final_salary_scheme_orig'].value_counts(normalize=True)

prev_scheme_counts = df['previous_final_salary_scheme_orig'].value_counts()
plt.figure(figsize=(8, 5))
bars = plt.bar(prev_scheme_counts.index, prev_scheme_counts.values, color=colours[:len(prev_scheme_counts)])
plt.title('Previous Final Salary Scheme Membership')
plt.xlabel('Previous Scheme')
plt.ylabel('Count')
for i, v in enumerate(prev_scheme_counts.values):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom', fontsize=10)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

#%%
#Public sector transfer club indicator
df['public_sector_transfer_club_pstc_transfer_in_indicator'].value_counts()
df['public_sector_transfer_club_pstc_transfer_in_indicator'].value_counts(normalize=True)

print("Missing values:", df['public_sector_transfer_club_pstc_transfer_in_indicator'].isna().sum())

#%%
#Previous final salary scheme if any
df['previous_final_salary_scheme_membership_if_any'].value_counts()
df['previous_final_salary_scheme_membership_if_any'].value_counts(normalize=True)

df['previous_final_salary_scheme_membership_if_any'] = df['previous_final_salary_scheme_membership_if_any'].fillna('N')
df['previous_final_salary_scheme_membership_if_any'].value_counts()
df['previous_final_salary_scheme_membership_if_any'].value_counts(normalize=True)

print("Missing values:", df['previous_final_salary_scheme_membership_if_any'].isna().sum())

prev_scheme_counts = df['previous_final_salary_scheme_membership_if_any'].value_counts()
plt.figure(figsize=(8, 5))
bars = plt.bar(prev_scheme_counts.index, prev_scheme_counts.values, color=colours[:len(prev_scheme_counts)])
plt.title('Previous Final Salary Scheme Membership (if any)')
plt.xlabel('Previous Scheme')
plt.ylabel('Count')
for i, v in enumerate(prev_scheme_counts.values):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom', fontsize=10)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

#%%
# County
county_counts = df['adm_ons_name'].value_counts(normalize=True).head(10)
print(county_counts)

plt.figure(figsize=(10, 6))
bars = plt.barh(county_counts.index[::-1], county_counts.values[::-1], color=colours[:len(county_counts)])
plt.title('Top 10 Counties by Member Count')
plt.xlabel('Number of Members')
plt.ylabel('County')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()


#%%
# Dates
# Convert and calculate age
df['dob'] = pd.to_datetime(df['date_of_birth_month'], errors='coerce')
reference_date = pd.to_datetime('2020-04-01')
df['age'] = (reference_date - df['dob']).dt.days / 365.25

# Drop nulls just in case
df_age = df.dropna(subset=['age'])

# Plot
plt.figure(figsize=(10, 6))
sns.histplot(data=df_age, x='age', kde=True, bins=30, color='#377eb8')
plt.title('Age Distribution of Members')
plt.xlabel('Age')
plt.ylabel('Number of Members')
plt.grid(True, linestyle='--', alpha=0.5)
#plt.show()

# Age by gender
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df_age, x='age', hue='gender', fill=True)
plt.title('Age Distribution by Gender')
plt.xlabel('Age')
plt.grid(True, linestyle='--', alpha=0.5)
#plt.show()

# Date joined scheme month
df['joined_scheme'] = pd.to_datetime(df['date_joined_scheme_month'], errors='coerce')
df['joined_year'] = df['joined_scheme'].dt.year

plt.figure(figsize=(10, 6))
sns.histplot(df['joined_year'].dropna(), bins=range(1970, 2025, 2), kde=False, color='#4daf4a')
plt.title('Year Joined Scheme')
plt.xlabel('Year')
plt.ylabel('Number of Members')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

# Date transition to 2015 scheme
df['transition_date'] = pd.to_datetime(df['date_of_transition_to_2015_scheme_month'], errors='coerce')
df['transition_year'] = df['transition_date'].dt.year

plt.figure(figsize=(10, 6))
sns.histplot(df['transition_year'].dropna(), bins=range(2014, 2021), color='#4daf4a')
plt.title('Transition to 2015 Scheme (by Year)')
plt.xlabel('Year')
plt.ylabel('Number of Members')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

# Date attains state pension
df['spa_date'] = pd.to_datetime(df['date_attains_state_pension_age_legislation_month'], errors='coerce')
df['spa_year'] = df['spa_date'].dt.year

plt.figure(figsize=(10, 6))
sns.histplot(df['spa_year'].dropna(), bins=range(2020, 2060, 2), color='#4daf4a')
plt.title('State Pension Age Attainment Year')
plt.xlabel('Year')
plt.ylabel('Number of Members')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()




#INDEPTH INSIGHTS

#%% Correlation Heatmap


# Copy and process DataFrame
df_heatmap = df.copy()

df_heatmap['part_time_flag'] = df_heatmap['part_time'].map({'N': 0, 'Y': 1})
reference_date = pd.to_datetime('2020-04-01')
df_heatmap['joined_scheme'] = pd.to_datetime(df_heatmap['date_joined_scheme_month'], errors='coerce')
df_heatmap['years_in_scheme'] = (reference_date - df_heatmap['joined_scheme']).dt.days / 365.25
df_heatmap['rank_encoded'] = pd.to_numeric(df_heatmap['rank'], errors='coerce')

# Clean variable selection
heatmap_data = df_heatmap[[
    'age',
    'actual_pay_over_12_months_final_salary_scheme',
    'part_time_flag',
    'years_in_scheme',
    'rank_encoded',
    'accrued_2015_scheme_pension_at_valuation_date',
    'x2019_20_employee_contribution_history'
]].replace([np.inf, -np.inf], np.nan).dropna()

heatmap_data = heatmap_data.loc[:, heatmap_data.std() > 0]
corr = heatmap_data.corr().dropna(axis=0, how='any').dropna(axis=1, how='any')

# Plot
plt.figure(figsize=(20, 18))
sns.heatmap(
    corr,
    annot=False, #no numbers in squares on heatmap as clutters visualisation
    cmap="coolwarm",
    linewidths=1.2,
    square=True,
    cbar_kws={'shrink': 0.6, 'label': 'Correlation'},
    mask=np.zeros_like(corr, dtype=bool)
)

plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(rotation=0, fontsize=10)
plt.title('Correlation Heatmap of Key Pension Determinants', fontsize=14, pad=20)
plt.tight_layout()
plt.show()

corr_pairs = corr.unstack().round(2).sort_values(ascending=False)
print(corr_pairs)





#%% Gender vs Pension Accrued - Box Plot



# Copy the original DataFrame
df_gender_pension = df[
    (df['accrued_2015_scheme_pension_at_valuation_date'] > 0)
].copy()

# Clean gender variable again if needed
df_gender_pension['gender'] = df_gender_pension['gender'].str.strip().str.lower()
df_gender_pension['gender'] = df_gender_pension['gender'].replace({
    'm': 'Male',
    'male': 'Male',
    'f': 'Female',
    'female': 'Female'
})

# Select relevant columns and drop missing pensions
df_gender_pension = df_gender_pension[['gender', 'accrued_2015_scheme_pension_at_valuation_date']].dropna()

# Plot
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=df_gender_pension,
    x='gender',
    y='accrued_2015_scheme_pension_at_valuation_date',
    palette='coolwarm'
)

plt.title('Pension Accrued by Gender', fontsize=14, pad=15)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Accrued 2015 Scheme Pension (£)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()





#%% Gender vs 2020 Pay - Box Plot (All Workers)

# Copy and clean
df_gender_pay = df[
    df['actual_pay_over_12_months_final_salary_scheme'] > 0
].copy()

# Clean gender
df_gender_pay['gender'] = df_gender_pay['gender'].str.strip().str.lower()
df_gender_pay['gender'] = df_gender_pay['gender'].replace({
    'm': 'Male',
    'male': 'Male',
    'f': 'Female',
    'female': 'Female'
})

# Select 2020 pay column
pay_col = 'actual_pay_over_12_months_final_salary_scheme'

# Drop missing values
df_gender_pay = df_gender_pay[['gender', pay_col]].dropna()

# Plot
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=df_gender_pay,
    x='gender',
    y=pay_col,
    palette='coolwarm'
)

plt.title('2020 Actual Pay by Gender (All Workers)', fontsize=14, pad=15)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Actual Pay Over 12 Months (£)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5) # Zoom in to focus on key middle values
plt.ylim(0, 100000)
plt.show()





#%% Gender vs 2020 Pay - Box Plot (Full-Time Workers Only)

# Filter full-time only
df_gender_pay_fulltime = df[
    (df['part_time'] == 'N') &
    (df['actual_pay_over_12_months_final_salary_scheme'] > 0)
].copy()

# Clean gender
df_gender_pay_fulltime['gender'] = df_gender_pay_fulltime['gender'].str.strip().str.lower()
df_gender_pay_fulltime['gender'] = df_gender_pay_fulltime['gender'].replace({
    'm': 'Male',
    'male': 'Male',
    'f': 'Female',
    'female': 'Female'
})

# Keep only full-time workers
df_gender_pay_fulltime = df_gender_pay_fulltime[
    (df_gender_pay_fulltime['part_time'] == 'N')
]

# Select and drop missing pay
df_gender_pay_fulltime = df_gender_pay_fulltime[['gender', pay_col]].dropna()

# Plot
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=df_gender_pay_fulltime,
    x='gender',
    y=pay_col,
    palette='coolwarm'
)

plt.title('2020 Actual Pay by Gender (Full-Time Workers Only)', fontsize=14, pad=15)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Actual Pay Over 12 Months (£)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.ylim(0, 100000) #upper limit I set to filter extreme anomalies and focus on key areas
plt.tight_layout()
plt.show()

# Import stats library
from scipy.stats import ttest_ind

# Group pay data by gender
male_pay = df_gender_pay_fulltime[df_gender_pay_fulltime['gender'] == 'Male'][pay_col]
female_pay = df_gender_pay_fulltime[df_gender_pay_fulltime['gender'] == 'Female'][pay_col]


# Filter to full-time workers with valid gender, age, and pension data
df_pension_age = df[
    (df['part_time'] == 'N') &
    (df['gender'].isin(['Male', 'Female'])) &
    (df['age'].notna()) &
    (df['accrued_2015_scheme_pension_at_valuation_date'].notna()) &
    (df['accrued_2015_scheme_pension_at_valuation_date'] > 0)
].copy()

# Plot
plt.figure(figsize=(12, 6))
sns.scatterplot(
    data=df_pension_age,
    x='age',
    y='accrued_2015_scheme_pension_at_valuation_date',
    hue='gender',
    alpha=0.5
)

# Add regression lines for each gender
sns.regplot(
    data=df_pension_age[df_pension_age['gender'] == 'Male'],
    x='age',
    y='accrued_2015_scheme_pension_at_valuation_date',
    scatter=False,
    color='blue',
    label='Male Trend'
)

sns.regplot(
    data=df_pension_age[df_pension_age['gender'] == 'Female'],
    x='age',
    y='accrued_2015_scheme_pension_at_valuation_date',
    scatter=False,
    color='orange',
    label='Female Trend'
)

plt.title('Pension Accrued vs Age (Full-Time Workers, by Gender)')
plt.xlabel('Age')
plt.ylabel('Accrued Pension (2015 Scheme)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()

# Refilter carefully
df_pension_age = df.copy()

# Clean gender and part-time
df_pension_age['gender'] = df_pension_age['gender'].str.strip().str.lower().replace({
    'm': 'Male', 'male': 'Male',
    'f': 'Female', 'female': 'Female'
})
df_pension_age['part_time'] = df_pension_age['part_time'].fillna('N')

# Filter for full-time and valid pension data
df_pension_age = df_pension_age[
    (df_pension_age['part_time'] == 'N') &
    (df_pension_age['age'].notna()) &
    (df_pension_age['accrued_2015_scheme_pension_at_valuation_date'].notna()) &
    (df_pension_age['accrued_2015_scheme_pension_at_valuation_date'] > 1000) #filter out anomalous lower-end results skewing graphs
]

# Now boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=df_pension_age,
    x='gender',
    y='accrued_2015_scheme_pension_at_valuation_date',
    palette='coolwarm'
)

plt.title('Accrued Pension (2015 Scheme) by Gender (Full-Time Workers Only)', fontsize=14)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Accrued Pension (£)', fontsize=12)
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.ylim(0,20000)
plt.tight_layout()
plt.show()





# Pension Accrued vs Years in Scheme (Full-Time Workers, by Gender)

# Create a copy of the original DataFrame
df_years_pension = df.copy()

# Filter for full-time workers only
df_years_pension = df_years_pension[df_years_pension['part_time'] == 'N']

# Calculate years in scheme
reference_date = pd.to_datetime('2020-04-01')
df_years_pension['joined_scheme'] = pd.to_datetime(df_years_pension['date_joined_scheme_month'], errors='coerce')
df_years_pension['years_in_scheme'] = (reference_date - df_years_pension['joined_scheme']).dt.days / 365.25

# Drop rows with missing key values and remove extreme pensions
df_years_pension = df_years_pension.dropna(subset=[
    'years_in_scheme',
    'accrued_2015_scheme_pension_at_valuation_date',
    'gender'
])
df_years_pension = df_years_pension[df_years_pension['accrued_2015_scheme_pension_at_valuation_date'] > 500]

# Plot regression lines
plt.figure(figsize=(12, 8))
sns.regplot(
    data=df_years_pension[df_years_pension['gender'] == 'Male'],
    x='years_in_scheme',
    y='accrued_2015_scheme_pension_at_valuation_date',
    scatter=False,
    label='Male',
    color='blue',
    lowess=True
)

sns.regplot(
    data=df_years_pension[df_years_pension['gender'] == 'Female'],
    x='years_in_scheme',
    y='accrued_2015_scheme_pension_at_valuation_date',
    scatter=False,
    label='Female',
    color='orange',
    lowess=True
)

plt.title('Pension Accrued vs Years in Scheme (Full-Time Workers, by Gender)', fontsize=14)
plt.xlabel('Years in Scheme')
plt.ylabel('Accrued Pension (£)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.ylim(bottom=0)
plt.tight_layout()
plt.show()

#Combine scatter graph with line graph - possible advanced visualisation based on GAD feedback
import statsmodels.api as sm

# Setup data
df_filtered = df.copy()
reference_date = pd.to_datetime('2020-04-01')
df_filtered['joined_scheme'] = pd.to_datetime(df_filtered['date_joined_scheme_month'], errors='coerce')
df_filtered['years_in_scheme'] = (reference_date - df_filtered['joined_scheme']).dt.days / 365.25

# Filter for full-time and valid values
df_filtered = df_filtered[
    (df_filtered['part_time'] == 'N') &
    (df_filtered['accrued_2015_scheme_pension_at_valuation_date'] > 0) &
    (df_filtered['years_in_scheme'] > 0)
]

# Limit to reasonable years
df_filtered = df_filtered[df_filtered['years_in_scheme'] <= 30]

# Set up plot
plt.figure(figsize=(12, 8))

# Scatter
sns.scatterplot(
    data=df_filtered,
    x='years_in_scheme',
    y='accrued_2015_scheme_pension_at_valuation_date',
    hue='gender',
    alpha=0.1,
    s=10,
    palette={'Male': 'blue', 'Female': 'orange'},
    edgecolor=None,
    legend=False
)

# Add LOWESS line for each gender
for gender, color in [('Male', 'blue'), ('Female', 'orange')]:
    subset = df_filtered[df_filtered['gender'] == gender]
    lowess = sm.nonparametric.lowess(
        subset['accrued_2015_scheme_pension_at_valuation_date'],
        subset['years_in_scheme'],
        frac=0.3
    )
    plt.plot(lowess[:, 0], lowess[:, 1], label=gender, color=color, linewidth=2)

# Labels and style
plt.title('Accrued 2015 Pension vs Years in Scheme\n(Full-Time Workers, by Gender)', fontsize=14)
plt.xlabel('Years in Pension Scheme', fontsize=12)
plt.ylabel('Accrued 2015 Pension (£)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()





#Pension vs Year Joined
# Ensure date is in datetime format and extract year
df['joined_scheme'] = pd.to_datetime(df['date_joined_scheme_month'], errors='coerce')
df['joined_year'] = df['joined_scheme'].dt.year

# Filter for valid pension values and full-time workers
df_filtered = df[
    (df['part_time'] == 'N') &
    (df['accrued_2015_scheme_pension_at_valuation_date'] > 0) &
    (df['joined_year'].notnull())
].copy()

# Create 5-year bins
start_year = int(df_filtered['joined_year'].min())
end_year = int(df_filtered['joined_year'].max())
bin_edges = list(range(start_year - (start_year % 5), end_year + 5, 5))
bin_labels = [f"{y}-{y+4}" for y in bin_edges[:-1]]

# Apply bins
df_filtered['join_cohort'] = pd.cut(df_filtered['joined_year'], bins=bin_edges, labels=bin_labels, right=False)

# Group and calculate average pension
grouped = df_filtered.groupby(['join_cohort', 'gender'])[
    'accrued_2015_scheme_pension_at_valuation_date'
].mean().reset_index()

# Drop NaN cohorts (if any)
grouped = grouped.dropna(subset=['join_cohort'])

# Plot
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=grouped,
    x='join_cohort',
    y='accrued_2015_scheme_pension_at_valuation_date',
    hue='gender',
    marker='o',
    palette={'Male': 'blue', 'Female': 'orange'}
)

plt.title('Average Accrued 2015 Pension by 5-Year Join Cohort\n(Full-Time Workers, by Gender)', fontsize=14)
plt.xlabel('Join Cohort (5-Year Intervals)')
plt.ylabel('Average Accrued 2015 Pension (£)')
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()




#Pay by rank
# Convert rank to numeric (cant remember if already done so to be safe)
df['rank'] = pd.to_numeric(df['rank'], errors='coerce')

# Filter for full-time, valid pay and ranks 80–82 - GAD asked for these ranks
df_rank_compare = df[
    (df['part_time'] == 'N') &
    (df['rank'].isin([80, 81, 82])) &
    (df['actual_pay_over_12_months_final_salary_scheme'].notna()) &
    (df['actual_pay_over_12_months_final_salary_scheme'] > 0)
]

# Confirm there is data to plot
print(df_rank_compare[['rank', 'gender', 'actual_pay_over_12_months_final_salary_scheme']].groupby(['rank', 'gender']).size())

# Create the boxplot
plt.figure(figsize=(12, 8))
sns.boxplot(
    data=df_rank_compare,
    x='rank',
    y='actual_pay_over_12_months_final_salary_scheme',
    hue='gender',
    showfliers=False
)
plt.title('Actual Pay by Rank and Gender (Ranks 80–82, Full-Time Workers)', fontsize=14)
plt.xlabel('Rank')
plt.ylabel('Actual Pay Over 12 Months (£)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()





#Pay over time by gender (same rank, full-time filtered) - keky asked for by GAD
# Create a copy of the original DataFrame
df_paytime = df.copy()

# Convert rank to numeric for filtering
df_paytime['rank'] = pd.to_numeric(df_paytime['rank'], errors='coerce')

# Filter for full-time workers ('N') and ranks 80, 81, 82
df_paytime = df_paytime[
    (df_paytime['part_time'] == 'N') &
    (df_paytime['rank'].isin([80, 81, 82]))
]

# Define pay columns and their corresponding years
pay_cols = {
    'x2015_rate_of_pensionable_pay_final_salary_scheme': 2015,
    'x2016_rate_of_pensionable_pay_final_salary_scheme': 2016,
    'x2017_rate_of_pensionable_pay_final_salary_scheme': 2017,
    'x2018_rate_of_pensionable_pay_final_salary_scheme': 2018,
    'x2019_rate_of_pensionable_pay_final_salary_scheme': 2019,
}

# Melt the DataFrame into long format
df_melted = df_paytime[['gender', 'rank'] + list(pay_cols.keys())].melt(
    id_vars=['gender', 'rank'],
    value_vars=pay_cols.keys(),
    var_name='year',
    value_name='pay'
)

# Map to actual years
df_melted['year'] = df_melted['year'].map(pay_cols)

# Remove missing or invalid values
df_melted = df_melted.dropna(subset=['pay'])
df_melted = df_melted[df_melted['pay'] > 0]

# Plot
plt.figure(figsize=(12, 7))
sns.lineplot(
    data=df_melted,
    x='year',
    y='pay',
    hue='gender',
    style='rank',
    markers=True,
    dashes=False
)

plt.title('Pensionable Pay Over Time by Gender and Rank (Ranks 80–82, Full-Time)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Rate of Pensionable Pay (£)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks([2015, 2016, 2017, 2018, 2019])
plt.legend(title='Gender / Rank')
plt.tight_layout()
plt.show()





#Promotions by age and gender
# Copy and clean
df_age_rank = df.copy()
df_age_rank['rank'] = pd.to_numeric(df_age_rank['rank'], errors='coerce')
df_age_rank['dob'] = pd.to_datetime(df_age_rank['dob'], errors='coerce')
reference_date = pd.to_datetime("2020-04-01")
df_age_rank['age'] = (reference_date - df_age_rank['dob']).dt.days / 365.25

# Filter
df_age_rank = df_age_rank[
    (df_age_rank['part_time'] == 'N') &
    (df_age_rank['rank'].between(80, 88)) &
    (df_age_rank['age'].notna())
]

# Plot: Boxplot to show distribution across ranks
plt.figure(figsize=(12, 6))
sns.boxplot(
    data=df_age_rank,
    x='rank',
    y='age',
    hue='gender',
    palette='Set1',
    fliersize=2
)
plt.title('Age Distribution by Rank and Gender (Full-Time Workers)', fontsize=14)
plt.xlabel('Rank')
plt.ylabel('Age')
plt.legend(title='Gender')
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()





# Age vs years of service

df_plot = df.copy()

# Rename columns for clarity - easier for me to produce interpretable graph
df_plot = df_plot.rename(columns={
    'pensionable_service_final_salary_scheme': 'years_of_service'
})

# Drop rows with missing values in required columns
df_plot = df_plot.dropna(subset=['age', 'years_of_service', 'gender'])

# Filter out any zero or negative service years if needed
df_plot = df_plot[df_plot['years_of_service'] > 0]

# Plotting
plt.figure(figsize=(12, 7))
sns.scatterplot(
    data=df_plot,
    x='years_of_service',
    y='age',
    hue='gender',
    alpha=0.4,
    palette={'Male': 'blue', 'Female': 'orange'},
    edgecolor=None,
    s=15
)

# Add LOWESS trend lines
sns.regplot(
    data=df_plot[df_plot['gender'] == 'Male'],
    x='years_of_service',
    y='age',
    scatter=False,
    lowess=True,
    color='blue',
    label='Male Trend'
)

sns.regplot(
    data=df_plot[df_plot['gender'] == 'Female'],
    x='years_of_service',
    y='age',
    scatter=False,
    lowess=True,
    color='orange',
    label='Female Trend'
)

plt.title('Age vs. Years of Service by Gender', fontsize=14)
plt.xlabel('Years of Service')
plt.ylabel('Age')
plt.legend(title='Gender')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()




#Average years of service by rank and gender
# Copy and clean data
df_service = df.copy()

# Ensure correct types
df_service['rank'] = pd.to_numeric(df_service['rank'], errors='coerce')
df_service['gender'] = df_service['gender'].astype(str)
df_service['pensionable_service_final_salary_scheme'] = pd.to_numeric(
    df_service['pensionable_service_final_salary_scheme'], errors='coerce')

# Filter only valid records
df_service = df_service[
    df_service['rank'].notna() &
    df_service['pensionable_service_final_salary_scheme'].notna()
]

# Group by rank and gender, and calculate mean years of service
df_grouped = df_service.groupby(['rank', 'gender'])[
    'pensionable_service_final_salary_scheme'].mean().reset_index()

# Plotting
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_grouped, x='rank', y='pensionable_service_final_salary_scheme', hue='gender', marker='o')

plt.title('Average Years of Service by Rank and Gender', fontsize=14)
plt.xlabel('Rank')
plt.ylabel('Average Years of Service')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()

print(df.columns.tolist())



```
